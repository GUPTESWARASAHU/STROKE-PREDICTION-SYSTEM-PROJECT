```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    "Logistic": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier()
}
```

 STEP 07: MODEL TRAINING

```python
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


model_predictions = {}
model_scores = {}


X_train_cleaned = np.nan_to_num(X_train, nan=0.0, posinf=np.finfo(np.float64).max, neginf=np.finfo(np.float64).min)
X_test_cleaned = np.nan_to_num(X_test, nan=0.0, posinf=np.finfo(np.float64).max, neginf=np.finfo(np.float64).min)

for name, model in models.items():
    print(f"\nTraining {name}...")
    try:
        model.fit(X_train_cleaned, y_train)
        y_pred = model.predict(X_test_cleaned)
        model_predictions[name] = y_pred


        accuracy = accuracy_score(y_test, y_pred)
        model_scores[name] = accuracy
        print(f"{name} Accuracy: {accuracy:.4f}")
        print(f"{name} Classification Report:\n{classification_report(y_test, y_pred)}")
        print(f"{name} Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")

    except Exception as e:
        print(f"Error training {name}: {e}")

print("\n--- Model Training Summary ---")
for name, accuracy in model_scores.items():
    print(f"{name}: Accuracy = {accuracy:.4f}")


model = models['Logistic']
model.fit(X_train_cleaned, y_train)
```
